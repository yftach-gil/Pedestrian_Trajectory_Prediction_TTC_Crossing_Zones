## Manipulate data for startified analysis

### remove rows with num_of_pairs=0 (leave only rows with existing ttc)

```python
for df in dfs:
    df.drop(df[df['num_of_pairs'] == 0].index, inplace=True)
    df.reset_index(drop=True, inplace=True)

len(df)
```

### Filter track to be in one of the ares (remover all outside one of the ares)

```python
for df in dfs:
    df.drop(
        df[
            (df['zone_zebra'] == 0) &
            (df['zone_1'] == 0) &
            (df['zone_2'] == 0) &
            (df['zone_3'] == 0)
        ].index,
        inplace=True
    )
    df.reset_index(drop=True, inplace=True)
```

```python
len(df)
```

### update the trackTotalNumFrames

```python
for df in dfs:
    # Get the maximum trackLifetime for each trackId
    track_durations = df.groupby('trackId')['trackLifetime'].max()
    track_durations=track_durations+1  # Add 1 to include the first '0' frame
    # Map the duration back to each row by trackId
    df['trackTotalNumFrames'] = df['trackId'].map(track_durations)
    # move to the 4th column
    column_to_move=df.pop("trackTotalNumFrames")
    df.insert(4, "trackTotalNumFrames", column_to_move)
```
