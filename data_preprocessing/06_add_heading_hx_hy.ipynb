```python
root_dir = r"C:\Users\yftac\Documents\00_Project_Afeka\ttc_calculation"
folder_name='05_data_with_pairs_folder'
```

```python
# Load dataframes
def load_dfs(file_name_list, root_dir, folder_name):
    os.chdir(root_dir)
    # change to data dir
    os.chdir(folder_name)
    dfs = []
    for i in range(len(file_name_list)):
        df=pd.read_csv(file_name_list[i])
        dfs.append(df)
    # change back to parent directory
    os.chdir('..')
    return dfs

dfs=load_dfs(file_name_data_with_pairs_list, root_dir, folder_name)
```

```python
# print dfs info
def print_dfs_info(dfs, file_name_data_with_pairs_list):
    for i in range(len(dfs)):
        print(f'{file_name_data_with_pairs_list[i]} df:\n')
        print(dfs[i].info())
        print('------------------------------------------------------')

print_dfs_info(dfs, file_name_data_with_pairs_list)
```

### Add Heading Vector hx, hy from heading in deg

```python
# function to calculate the direction vector
def heading_deg_to_vector(heading_deg):
    c=1     # assume this is a unit vector
    hx=c*np.cos(np.radians(heading_deg))
    hy=c*np.sin(np.radians(heading_deg))
    return hx, hy
```

```python
dfs_temp=dfs

for x in range(len(dfs_temp)):
    print('df name:' , file_name_data_with_pairs_list[x].replace('_data_with_pairs.csv', '') )
    df_temp=dfs_temp[x]
    df_temp['hx'] = None
    df_temp['hy'] = None
# add the vector to the df hx and hy columns    
    for i in tqdm(range(len(df_temp)), desc='Adding heading vectors'):
        hx_temp, hy_temp = heading_deg_to_vector(df_temp.iloc[i]['heading'])
        df_temp.loc[df_temp.index[i], 'hx'] = hx_temp
        df_temp.loc[df_temp.index[i], 'hy'] = hy_temp

    dfs[x]=df_temp
    print('------------------------------------------------------')
print('finished adding heading vectors')
```

```python
recordingId_list = file_name_data_with_pairs_list 
for i in range(len(recordingId_list)):
    recordingId_list[i]=recordingId_list[i].replace('_data_with_pairs.csv', '')
recordingId_list
```

```python
### save to csv files  ###
os.chdir(root_dir)

def save_dfs(dfs, recordingId_list):
    # create folder to save files 
    folder_name='06_data_with_heading_vectors_folder'
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    # change to working dir
    os.chdir(folder_name)
    
    # initialize a list of the files to return by the function
    file_name_data_with_heading_vectors_list=[]
    
    # iterate dfs and save
    for x in tqdm( range(len(dfs)) , desc="Saving CSV files" ):
        # save as CSV file
        file_name=str(recordingId_list[x]) + '_data_with_heading_vectors' + '.csv'
        dfs[x].to_csv(file_name, index=False)
        # append the file name to the list
        file_name_data_with_heading_vectors_list.append(file_name)
        
    print(f"files saved to: {os.getcwd()}")
    # change back to parent directory
    os.chdir('..')

    return file_name_data_with_heading_vectors_list

file_name_data_with_heading_vectors_list = save_dfs(dfs, recordingId_list)
```
