## Create pairs of pedestrian-vehicle

Assume that bycicle have no affect on the pedestrian-car/truck behaiviour so they are ignored

```python
def create_pairs_column(df, df_name):
    # add df column that will contain the pairs information
    df['pairs']=[[] for _ in range(len(df))]  # initialize with empty list

    # iterate on all frames of df
    for f in tqdm(range(df['frame'].min(), df['frame'].max()+1, 1), desc=f'Adding pairs: {df_name}'):
        # print(f'------------ frame={f} ------------')
        pairs_list=[]   # the list will intialize for each frame
        # find rows for frame
        row_indexes_of_frame = df.index[df['frame'] == f].tolist()
           
        # create dictionary with key=row index, value=class, the dictionary contains all objects+rows in the frame
        frame_dict=dict.fromkeys(row_indexes_of_frame)
        for r in row_indexes_of_frame:
            frame_dict[r]=df.iloc[r]['class']
        # print('frame_dict\n', frame_dict, '\n')
        
        # find pairs of pedestrian and cars or truck_bus
        for key, value in frame_dict.items():
            if value == 'pedestrian':
                for key2, value2 in frame_dict.items():
                    if value2 == 'car' or value2 == 'truck_bus':
                        # print(f'({key},pedestrian) and ({key2}, {value2})') # print each pair found                                            
                        df.at[key, 'pairs'].append([key2, value2])          # add to the row['pairs'] as list 
    return df
```

```python
def add_num_of_pairs_column(df, df_name):
    df['num_of_pairs']=[0 for _ in range(len(df))]  # initialize column with 0
    
    for i in tqdm(range(len(df)), desc='Adding num_pairs: {df_name}'):
        df.at[i, 'num_of_pairs']=len(df.at[i, 'pairs'])
    return df
```

```python
# process all dfs
for x in range(len(dfs)):
    print(f'-------- df name: {file_name_filtered_with_dim_data_list[x]} --------')
    print('total number of rows:', dfs[x].shape[0], '\n')
    df_name= file_name_filtered_with_dim_data_list[x]
    dfs[x]=create_pairs_column(dfs[x], df_name)
    dfs[x]=add_num_of_pairs_column(dfs[x], df_name)
    print()
print('Finished')
```

```python
### save to csv files  ###
os.chdir(root_dir)

def save_dfs(dfs, recordingId_list):
    # create folder to save files 
    folder_name='05_data_with_pairs_folder'
    if not os.path.exists(folder_name):
        os.makedirs(folder_name)
    # change to working dir
    os.chdir(folder_name)
    
    # initialize a list of the files to return by the function
    file_name_data_with_pairs_list=[]
    
    # iterate dfs and save
    for x in tqdm(range(len(dfs)), desc="Saving CSV files"):
        # save as CSV file
        file_name=str(recordingId_list[x]) + '_data_with_pairs' + '.csv'
        dfs[x].to_csv(file_name, index=False)
        # append the file name to the list
        file_name_data_with_pairs_list.append(file_name)
        
    print(f"files saved to: {os.getcwd()}")
    # change back to parent directory
    os.chdir('..')

    return file_name_data_with_pairs_list

file_name_data_with_pairs_list=save_dfs(dfs, recordingId_list)
```
